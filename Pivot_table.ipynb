```python
titanic.groupby('sex')[['survived']].sum()
```

```python
titanic.groupby('sex')[['survived']].mean()*100
```

```python
titanic.groupby(['sex','class'])['survived'].aggregate('sum').unstack()
```

```python
titanic.groupby(['sex','class'])['survived'].sum().unstack()
```

```python
titanic.pivot_table('survived', columns='sex')
```

```python
titanic.pivot_table('survived',index='sex', columns='class')
```

```python
age = pd.cut(titanic['age'],[0,30,50,80])
titanic.pivot_table('survived',index=['sex',age,'who'], columns='class')
```

```python
fare = pd.qcut(titanic['fare'],3)
print(fare)
```

```python
titanic.pivot_table('survived',['sex',age,'who'], [fare,'class'])
```

```python
titanic.pivot_table(index='sex', columns='class',
                   aggfunc={'survived':sum,
                           'fare':'min',
                           'who':'count'})
```

```python
titanic.pivot_table('survived',
                    index='sex', 
                    columns='class',margins=True)
```
